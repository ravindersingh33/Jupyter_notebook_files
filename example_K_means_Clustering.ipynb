```python
data_clust = data[['Player','Ave','SR']]
data_clust.head()
```

```python
df_scalaed = data_clust[['Ave','SR']]
df_scalaed.head()
```

```python
# rescaling
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
scaler =  StandardScaler()
scaled_data = scaler.fit_transform(df_scalaed)
scaled_data
```

```python
df_Scaled = pd.DataFrame(scaled_data)
df_Scaled.columns = ['Ave','SR']
df_Scaled
```

```python
from sklearn.cluster import KMeans
```

```python
# modelling

kmeans = KMeans(n_clusters = 4,random_state=100)
kmeans.fit(df_Scaled)
```

```python
kmeans.labels_
```

```python
data_clust.head()
```

```python
data_clust['cluster_id'] = kmeans.labels_
data_clust.head()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
# complete linkage 
mergings = linkage(df_Scaled, method ='complete', metric ='euclidean')
dendrogram(mergings)
plt.show()
```

```python
# 4 clusters 
cluster_label = cut_tree(mergings,n_clusters =4)
cluster_label
```

```python
cluster_label.shape
```

```python
cluster_label = cluster_label.reshape(-1,)
```

```python
cluster_label.shape
```

```python
data_clust['H_cluster'] = cluster_label
data_clust.head()
```

```python
data_clust.H_cluster.value_counts()
```
